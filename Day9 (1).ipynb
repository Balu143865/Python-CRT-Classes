```python
# Its a example of classes 
class Dog:
    def __init__(self, name, color, size):         # Attributes name, color, size
        self.name = name
        self.color = color
        self.size = size

    def walking(self, dist):                                         # Methods are walking, sleeping, eating
        self.dist = dist
        print(f"I am {self.name} and I am walked {self.dist}KM.")

    def sleeping(self):
        print(f"I am {self.name} and I am sleeping.")

    def eating(self):
        print(f"I am {self.name} and I am eating after walking {self.dist}KM.")


dog1 = Dog("Juile", "Brown-black", "2.5ft")
dog1.walking(40)
dog1.sleeping()
dog1.eating()

print("*" * 40)

dog2 = Dog("bruno", "gold", "2ft")
dog2.walking(20)
dog2.sleeping()
dog2.eating()

print("*" * 40)

dog3 = Dog("orio", "white", "3ft")
dog3.walking(30)
dog3.sleeping()
dog3.eating()

# print(dog1.name)
# print(dog2.name)
```

```python
class ATM:
    def __init__(self, account_no, name, balance = 0):
        self.account_no = account_no
        self.name = name
        self.balance = balance
        self.history = ""

    def set_pin(self, pin):
        self.pin = pin
        print(f"Your pin is successfully created")

    def change_pin(self, old_pin, new_pin):
        if self.pin == old_pin:
            self.pin = new_pin
            print(f"Your pin is successfully updated")
        else:
            print(f"Enter the correct pin")

    def deposit(self, amount, pswd):
        if self.pin == pswd:
            self.balance += amount
            self.history += f"Amount {amount} Credited" + "\n"
            print(f"Amount {amount} RS is Credited! and your available balance is {self.balance} Rs")
        else:
            print(f"Check the password you entered")
        
    
    def withdraw(self, amount, pswd):
        if self.pin == pswd:
            if self.balance > amount:
                self.balance -= amount
                self.history += f"Amount {amount} Debited" + "\n" 
                print(f"{amount} is Debited! and your available balance {self.balance} Rs")
            else:
                print(f"Insufficient balance")
        else:
            print(f"Invalid password")

    def check_balance(self, pswd):
        if pswd == self.pin:
            print(f"Your balance is {self.balance} Rs")
        else:
            print(f"Invalid password")

    def summary(self, pswd):
        if pswd == self.pin:    
            print(f"""Name: {self.name}
Account number: {self.account_no}
Available balance: {self.balance}""")

    def trans_history(self, pswd):
        if self.pin == pswd:
            print(self.history)
        else:
            print(self.history) 
 

user1 = ATM("123456", "Balu")
user1.set_pin(123)
user1.deposit(5000, 123)
user1.withdraw(2000, 123)
user1.check_balance(123)
user1.summary(123)
print("*" * 40)
user1.trans_history(123)


        
```

```python
# Program 2
class ATM:
    def __init__(self, account_no, name, balance = 0):
        self.account_no = account_no
        self.name = name
        self.balance = balance
        self.history = ""

    def isPassword(self, string):
        if len(string) >= 10:
            sc = 0
            uc = 0
            lc = 0
            dc = 0
            for char in string:
                if char .isupper():
                    uc += 1
                elif char.islower():
                    lc += 1
                elif not char.isalnum():
                    sc += 1
                else:
                    dc += 1
            if uc >= 2 and lc >= 2 and dc >= 2 and sc >= 2:
                return True
        return False
                    

    def set_pin(self, pin):
        # self.pin = pin
        if self.isPassword(str(pin)):
            self.pin = pin
            print(f"Your pin is successfully created")
        else:
            print(f"Recreate the pin!")

    def change_pin(self, old_pin, new_pin):
        
        if self.isPassword(str(new_pin)) and self.pin == old_pin:
            self.pin = new_pin
            print(f"Your pin is successfully updated")
        else:
            print(f"Enter the correct pin")

    def deposit(self, amount, pswd):
        if self.pin == pswd:
            self.balance += amount
            self.history += f"Amount {amount} Credited" + "\n"
            print(f"Amount {amount} RS is Credited! and your available balance is {self.balance} Rs")
        else:
            print(f"Check the password you entered")
        
    
    def withdraw(self, amount, pswd):
        if self.pin == pswd:
            if self.balance > amount:
                self.balance -= amount
                self.history += f"Amount {amount} Debited" + "\n" 
                print(f"{amount} is Debited! and your available balance {self.balance} Rs")
            else:
                print(f"Insufficient balance")
        else:
            print(f"Invalid password")

    def check_balance(self, pswd):
        if pswd == self.pin:
            print(f"Your balance is {self.balance} Rs")
        else:
            print(f"Invalid password")

    def summary(self, pswd):
        if pswd == self.pin:    
            print(f"""Name: {self.name}
Account number: {self.account_no}
Available balance: {self.balance}""")

    def transaction_history(self, pswd):
        if self.pin == pswd:
            print(self.history)
        else:
            print(self.history) 
 

user1 = ATM("123456", "Balu")
user1.set_pin("Balu@123")
user1.set_pin("Balu@12345A!")
user1.deposit(5000, "Balu@12345A!")
user1.withdraw(2000, "Balu@12345A!")
user1.check_balance("Balu@12345A!")
user1.transaction_history("Balu@12345A!")
user1.summary("Balu@12345A!")
print("*" * 40)
user1.transaction_history("Balu@12345A!")


        
```

```python
class library_management:
    def  __init__(self):
        self.in_store = []
        self.user_list = []

    def new_book(self, book):
        self.in_store.append(book)
        print(f"The book '{book.upper()}' is successfully added")

    def issue_book(self, name, book):
        for i in self.user_list:
            if i.name == name and book in self.in_store:
                i.record.append(book)
                print(f"Book issued successfully!")
                break
        else:
            print(f"User not found")
            
    def register(self, name, mobile_no):
        if user in self.user_list:
            print(f"User already exist!")
        else:
            user1 = user(name, mobile_no)
            self.user_list.append(user1)
            print(f"User reigistration successful!")

    def book_return(self, name, book):
        for i in self.user_list:
            if i.name == name:
                i.record.remove(book)
                self.in_store.append(book)
                print(f"Book return successful!")

    def details(self, name):
        for i in self.user_list:
            if i.name == name:
                print(f"""Name:{i.name}
Contact: {i.mobile_no}
Books he has: {", ".join(i.record)} """)
        
        


class user:
    def __init__(self, name, mobile_no):
        self.name = name
        self.mobile_no = mobile_no
        self.record = []


lib1 = library_management()
books = ["Maths", "phys", "social", "Tel", "chem"]

for i in books:
    lib1.new_book(i)
print("*" * 40)    
user_details = {
    "Balu": "6304045279",
    "Anjali" : "75692",
    "Ajay" : "idrbhgvus",
    "Aira" : "jksnvhsnfvbshfuy"
}
for name, mobile in user_details.items():
    lib1.register(name, mobile)

print("*" * 40)

lib1.issue_book("Balu", "Maths")
lib1.issue_book("Balu", "Tel")
lib1.issue_book("Balu", "social")
lib1.issue_book("Balu", "chem")

print('*' * 40)
lib1.details("Balu")

print("*" * 40)
lib1.book_return("Balu","Maths")

print("*" * 40)
lib1.details("Balu")

```
