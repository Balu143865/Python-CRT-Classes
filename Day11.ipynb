# Stack in Python 

```python
class Stack:

    def __init__(self):
        self.stack_list = []
        self.length = 0

    def push(self, value):
        self.stack_list.append(value)
        self.length += 1
        print(f"Value {value} pushed successfully")

    def pull(self):
        popped = self.stack_list.pop()
        self.length -= 1
        print(f"Value {popped} is pulled successfully")

    def length(self):
        return len(self.stack_list)

    def peek(self):
        return self.stack_list[-1]

    def max(self):
        return max(self.stack_list)

    def display(self):
        for ele in self.stack_list[::-1]:
            print(f"""{ele}
|""")


s1 = Stack()
s1.push(10)
s1.push(20)
s1.push(30)
s1.push(40)
s1.push(50)
print("*" * 40)

s1.pull()
s1.display()
print("*" * 40)

print(s1.peek())

s1.display()
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class Stacks:
    def __init__(self):
        self.length = 0
        self.head = None
        self.tail = None

    def push(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length += 1

    def pull(self):
        if self.head is None:
            print(f"The stack is empty!")
        else:
            current = self.head
            while current.next != self.tail:
                current = current.next
            current.next = None
        self.length -= 1

    def peek(self):
        return self.tail.value

    def length(self):
        return self.length

    def max(self):
        maxi = float("-inf")
        current = self.head
        while current:
            if current.value > maxi:
                maxi = current.value
            current = current.next
        return maxi


    def __str__(self):
        values = []
        current = self.head
        while current:
            values.append(str(current.value))
            current = current.next
        return """\n|\n""".join(values[::-1])


s1 = Stacks()
elements = [10,20,30,40,50]

for i in elements:
    s1.push(i)

print(s1)
print("*" * 40)
print(s1.peek())
print("*" * 40)
print(f"The maxi is {s1.max()}")
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.length = 0   # count store cheyadaniki

    def append(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:   # last node ki velle varaku
                current = current.next
            current.next = new_node
        self.length += 1   # new node add aina ppudu count perugutundi

    def count_nodes(self): 
        return self.length

    def display(self):
        current = self.head
        while current:
            print(current.value, end=" -> ")
            current = current.next
        print("None")


string = input("Enter a string: ")

ll = LinkedList()

for ch in string:   # each character ni append cheyyadam
    ll.append(ch)

print("Linked List contents: ")
ll.display()

print("Total count of characters:", ll.count_nodes())
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length += 1

    def printing(self):
        current = self.head
        while current:
            print(current.value, end = "-> ")
            current = current.next
        print()

    def reverse(self):
        if self.head is None:
            print(f"the list is empty")
        else:
            prev = None
            current = self.head
            while current:
                next_node = current.next
                current.next = prev
                prev = current
                current = next_node
            self.head = prev

link1 = LinkedList()
array = [10,20,30,40,50,60]

for i in array:
    link1.append(i)

link1.printing()
print("*" * 40)
link1.reverse()
link1.printing()
```
